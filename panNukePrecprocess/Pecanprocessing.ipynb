```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import sys
```

```python
masks = np.load("/home/ghulam/Downloads/masks.npy")
```

```python
for ids,instances in enumerate(masks):
    print("Mask{} Start ".format(ids))
    channel0 = instances[:,:,0]
    channel1 = instances[:,:,1]
    channel2 = instances[:,:,2]
    channel3 = instances[:,:,3]
    channel4 = instances[:,:,4]
    instanceslist = [channel0,channel1,channel2,channel3,channel4]
    outputs = np.zeros(shape=(256,256),dtype=np.int16)
    for idx,channal in enumerate(instanceslist):
        #print("channel{} Start ".format(idx))
        for (i,j) in zip(np.nditer(channal),np.nditer(outputs,op_flags=['readwrite'])):
            #print(i)
            #plt.pause(0.2)
            #print(j)
            if i == 0 and j != 0 :
                j[...] = j + i
            elif j == 0 and i != 0:
                j[...] = j + i
            elif i != 0 and j != 0 :
                j[...] = j
            else:
                j[...] = 0
            
    outputs = contagious_labeling(outputs,False)
    
    print("instance Mask{}  completed".format(ids))
    channel0[channel0 > 0] = 1
    channel1[channel1 > 0] = 2
    channel2[channel2 > 0] = 3
    channel3[channel3 > 0] = 4
    channel4[channel4 > 0] = 5
    classResultsArray = np.zeros(shape=(256,256),dtype=np.int16)
    for id_,channals in enumerate(instanceslist):
        #print("channel{} for class Start".format(id_))
        for (i,j) in zip(np.nditer(channals),np.nditer(classResultsArray,op_flags=['readwrite'])):
            #print(i,j)
            if i == 0 and j != 0:
                j[...] = j + i
            elif i != 0 and j == 0:
                j[...] = j + i
            elif i != 0 and j !=0 :
                j[...] = j
            else:
                j[...] = 0
            
        #print("channel{} completed".format(id_))
    print("Class Mask{}  completed".format(ids))
    annotation = np.dstack([outputs,classResultsArray])    
    print("final array formed ",annotation.shape)
    np.save("/home/ghulam/Desktop/panNuke/Labels/img_{}.npy".format(ids),annotation)
```

```python
def contagious_labeling(pred, by_size=False):
    """
    Rename all instance id so that the id is contiguous i.e [0, 1, 2, 3] 
    not [0, 2, 4, 6]. The ordering of instances (which one comes first) 
    is preserved unless by_size=True, then the instances will be reordered
    so that bigger nucler has smaller ID
    Args:
        pred    : the 2d array contain instances where each instances is marked
                  by non-zero integer
        by_size : renaming with larger nuclei has smaller id (on-top)
    """
    pred_id = list(np.unique(pred))
    pred_id.remove(0)
    if len(pred_id) == 0:
        return pred # no label
    if by_size:
        pred_size = []
        for inst_id in pred_id:
            size = (pred == inst_id).sum()
            pred_size.append(size)
        # sort the id by size in descending order
        pair_list = zip(pred_id, pred_size)
        pair_list = sorted(pair_list, key=lambda x: x[1], reverse=True)
        pred_id, pred_size = zip(*pair_list)

    new_pred = np.zeros(pred.shape, np.int32)
    for idx, inst_id in enumerate(pred_id):
        new_pred[pred == inst_id] = idx + 1    
    return new_pred
```
